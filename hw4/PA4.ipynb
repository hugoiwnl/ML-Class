````python
import csv
from collections import defaultdict

file_path = './data/movies.dat'

def load_data(file_path):
    with open(file_path, 'r') as file:
        reader = csv.reader(file, delimiter='\t')  # Adjust delimiter based on your data
        data = [row[0].strip().split('::') for row in reader]
    return data

movies_df = load_data('./data/movies.dat')
users_df = load_data('./data/users.dat')
ratings_df = load_data('./data/ratings.dat')
````

All ratings are contained in the file **ratings.dat** and are in the following format:

UserID - MovieID - Rating - Timestamp

=====================================

User information is in the file **users.dat** and is in the following format:

UserID - Gender - Age - Occupation - Zip-code

=====================================

Movie information is in the file **movies.dat** and is in the following format:

MovieID - Title - Genres

### Creating feature vector
So for the final feature vector, I would like to group the ratings and users by UserID, and then I would like to add a feature for every genre possible and have a 1 or a 0 depending on the specific movie of the rating for every genre. My final feature would look like this:

UserId - MovieId - Gender - Age - Occupation - Genres

Where genres would have 18 new features (listed below):

````python
genres = set()
for movie in movies_df:
    genres_tmp = movie[2].split("|")
    for genre in genres_tmp:
        genres.add(genre)
print(f"There are {len(genres)} genres.")
print("Those are: ")
print()
for g in genres:
    print(g)
````

Now I am making a dictionary where the keys are movieId's and values are lists of numerical genres transformation which I will use after for feature creatinng.

````python
movies_genres = defaultdict(list)
for movie in movies_df:
    genres_tmp = []
    for g in genres:
        if g in movie[2]:
            genres_tmp.append(1)
        else:
            genres_tmp.append(0)
    movies_genres[movie[0]] = genres_tmp
````

Now I will create the independent variable X and the dependant Y.

````python
X = []
Y = []

for rating in ratings_df:
    # Creating a list to append to X
    row = []
    
    # Adding userID, movieID
    row.append(int(rating[0]))
    row.append(int(rating[1]))

    # Adding user info (gender, age, occupation)
    user = users_df[int(rating[0]) - 1]
    # If user is male, I append 1, if user is female I append 0
    if(user[1] == 'M'):
        row.append(1)
    else:
        row.append(0)
    row.append(int(user[2]))
    row.append(int(user[3]))

    # Adding genres info
    row_final = row + movies_genres[rating[1]]

    # At the end I append this whole row to the X
    X.append(row_final)

    # I also append the rating to the target variable
    Y.append(int(rating[2]))
````

Now I need to remove users with less than 100 reviews.

````python
from collections import defaultdict

# Constructing a dictionary with ratings count for each user
num_ratings = defaultdict()

for rating in X:
    if(str(rating[0]) not in num_ratings.keys()):
         num_ratings[str(rating[0])] = 1
    else:
        num_ratings[str(rating[0])] += 1

# Making a table of all users to be deleted
users_to_delete = []

for user, num in num_ratings.items():
    if(num < 100):
        users_to_delete.append(user)


# Filtering ratings dataframe by deleting ratings of users with less than 100 ratings
X_filtered = []
Y_filtered = []

for index in range(len(X)):
    if(str(X[index][0]) in users_to_delete):
        continue
    else:
        X_filtered.append(X[index])
        Y_filtered.append(Y[index])
````

````python
print(f"Number of ratings removed: {len(X) - len(X_filtered)}.")
````

The last thing is splitting the data into training and test data.

````python
X_test = []
X_train = []

y_test = []
y_train = []

for index in range(len(X_filtered)):
    if(int(X_filtered[index][0]) < 1001):
        X_test.append(X_filtered[index])
        y_test.append(Y_filtered[index])
    else:
        X_train.append(X_filtered[index])
        y_train.append(Y_filtered[index])
````

Evaluate the best classifiers from the previous task in more detail (the best SVM and

confusion matrix for the classifier .
Don’t use scikit’s functions for this purpose but rather implement this functionality

````python
def compute_confusion_matrix(actual, predicted):
    classes = sorted(set(actual))  # Get unique classes in the data
    num_classes = len(classes)
    
    # Initialize the confusion matrix with zeros
    confusion_matrix = [[0] * num_classes for _ in range(num_classes)]
    
    # Populate the confusion matrix
    for i in range(len(actual)):
        actual_class = actual[i]
        predicted_class = predicted[i]
        
        actual_index = classes.index(actual_class)
        predicted_index = classes.index(predicted_class)
        
        confusion_matrix[actual_index][predicted_index] += 1
    
    return confusion_matrix

def print_confusion_matrix(confusion_matrix, classes):
    # Print the confusion matrix
    print("Confusion Matrix:")
    
    # Print header row
    header = ["Act/Pred"] + classes
    print("{:<20}".format(" "), end="")
    for item in header:
        print("{:<10}".format(item), end="")
    print("\n" + "-" * 80)
    
    # Print matrix
    for i in range(len(confusion_matrix)):
        actual_class = classes[i]
        row = [actual_class] + confusion_matrix[i]
        
        print("{:<20}".format(actual_class), end="")
        for item in row:
            print("{:<10}".format(item), end="")
        print()

# Example usage:
# Assuming svm_predictions and mlp_predictions are your predicted labels
svm_confusion_matrix = compute_confusion_matrix(y_test, svm_predictions)
mlp_confusion_matrix = compute_confusion_matrix(y_test, mlp_predictions)

print_confusion_matrix(svm_confusion_matrix, classes=sorted(set(y_test)))
print_confusion_matrix(mlp_confusion_matrix, classes=sorted(set(y_test)))
````

The MLP confusion matrix looks much better, having a lot of different predictions, but we can see that it is not doing such a good job, where it makes a lot of mistakes, in the first and the second classes more mistakes than the correct predictions even, so those classes seem to be hard to predict for our model. \
I will now make a function that computes the accuracties of the model according to their confusion matrix.

````python
def compute_accuracy(confusion_matrix):
    correct_predictions = sum(confusion_matrix[i][i] for i in range(len(confusion_matrix)))
    total_samples = sum(sum(row) for row in confusion_matrix)

    accuracy = correct_predictions / total_samples
    return accuracy

# Assuming confusion_matrix is your 2D confusion matrix
svm_accuracy = compute_accuracy(svm_confusion_matrix)
mlp_accuracy = compute_accuracy(mlp_confusion_matrix)
print("SVM Accuracy:", svm_accuracy)
print("MLP Accuracy:", mlp_accuracy)
````

## Subtask 2: Forward Propagation
Implement forward propagation for an MLP classifier with two hidden layers and ReLU 
activation function. Use small hidden layers, each containing 10 neurons. The ML 
must have a single output. Initialize weights randomly as discussed in class. Initiali e
biases as all-zero vectors.

````python
import numpy as np

def initialize_parameters(input_size, hidden_size1, hidden_size2, output_size):
    np.random.seed(42)  # for reproducibility
    W1 = np.random.randn(input_size, hidden_size1)
    b1 = np.zeros(hidden_size1)
    W2 = np.random.randn(hidden_size1, hidden_size2)
    b2 = np.zeros(hidden_size2)
    W3 = np.random.randn(hidden_size2, output_size)
    b3 = np.zeros(output_size)

    parameters = {"W1": W1, "b1": b1, "W2": W2, "b2": b2, "W3": W3, "b3": b3}

    return parameters

def forward_propagation(X, parameters, hidden_activation, output_activation):
    W1 = parameters['W1']
    b1 = parameters['b1']
    W2 = parameters['W2']
    b2 = parameters['b2']
    W3 = parameters['W3']
    b3 = parameters['b3']

    # Layer 1
    Z1 = np.dot(X, W1) + b1
    A1 = hidden_activation(Z1)

    # Layer 2
    Z2 = np.dot(A1, W2) + b2
    A2 = hidden_activation(Z2)

    # Output layer
    Z3 = np.dot(A2, W3) + b3
    A3 = output_activation(Z3)

    # Cache intermediate values for backward propagation
    cache = (Z1, A1, W1, b1, Z2, A2, W2, b2, Z3, A3, W3, b3)

    return A3, cache


# Defining the activation function
def relu(Z):
    return np.maximum(0, Z)

# The parameters
input_size = 2
hidden_size1 = 10
hidden_size2 = 10
output_size = 1

# These are the pairs I use for computing outputs
input_pairs = x_train

# Initializing parameters
parameters = initialize_parameters(input_size, hidden_size1, hidden_size2, output_size)


# Forward propagation for each input pair
outputs = []
for X in input_pairs:
    output, _ = forward_propagation(X, parameters, relu, relu)
    outputs.append(output)
````

## Subtask 3: Finite Difference Gradient & Optimization

````python
import h5py
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the data
with h5py.File('classification.h5', 'r') as hf:
    X_train = np.array(hf.get('x_train'))
    y_train = np.array(hf.get('y_train'))
    X_test = np.array(hf.get('x_test'))
    y_test = np.array(hf.get('y_test'))

# Normalize the features
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
````

First I start by defining a sigmoid activation function which I use for my output layer since we are dealing with binary classification, then I define a function for initializing weights and biases according to the architecture of the NN, after that I define forward propagation function as well as mean perceptron loss. At the end I compute the gradient approximation for weights and biases, I define a function for accuracy evaluation and a function for visualizing the results.

````python
import numpy as np
from _stochastic_optimizers import AdamOptimizer
import time
import matplotlib.pyplot as plt

# Another activation function to use
def sigmoid(Z):
    return 1 / (1 + np.exp(-Z))

# Another function for initialization of weights and biases for this task
def initialize_weights_biases(input_size, hidden_layer_num, hidden_layer_size, output_size):

    # Initializing the arrays for weights and biases
    weights = []
    biases = []

    # Adding the weights and biases for the first layer
    weights.append(np.random.randn(input_size, hidden_layer_size) * 0.01)
    biases.append(np.zeros(hidden_layer_size))

    # For the hidden layers
    for i in range(hidden_layer_num - 1):
        weights.append(np.random.randn(hidden_layer_size, hidden_layer_size) * 0.01)
        biases.append(np.zeros(hidden_layer_size))

    # For the output layer
    weights.append(np.random.randn(hidden_layer_size, output_size) * 0.01)
    biases.append(np.zeros(output_size))

    return weights, biases

# Forward propagation
def forward_propagation(X, weights, biases, hidden_activation, output_activation):
    A = X
    for W, b in zip(weights[:-1], biases[:-1]):
        Z = A.dot(W) + b
        A = hidden_activation(Z)
    Z = A.dot(weights[-1]) + biases[-1]
    output_prediction = output_activation(Z)
    return output_prediction

# Mean perceptron loss
def mean_perceptron_loss(output_prediction, y):
    loss = np.mean((y - output_prediction) ** 2)
    return loss

# Gradient computation
def compute_gradient(X, y, weights, biases, epsilon=1e-7):
    gradients = []
    gradients_biases = []

    for l in range(len(weights)):
        W_grad = np.zeros(weights[l].shape)
        b_grad = np.zeros(biases[l].shape)

        # Gradients for weights
        for i in range(weights[l].shape[0]):
            for j in range(weights[l].shape[1]):
                W_plus = np.copy(weights[l])
                W_plus[i, j] += epsilon
                weights_plus = weights[:]
                weights_plus[l] = W_plus
                output_plus = forward_propagation(X, weights_plus, biases, relu, sigmoid)
                loss_plus = mean_perceptron_loss(output_plus, y)

                W_minus = np.copy(weights[l])
                W_minus[i, j] -= epsilon
                weights_minus = weights[:]
                weights_minus[l] = W_minus
                output_minus = forward_propagation(X, weights_minus, biases, relu, sigmoid)
                loss_minus = mean_perceptron_loss(output_minus, y)

                W_grad[i, j] = (loss_plus - loss_minus) / (2 * epsilon)

        # Gradients for biases
        for i in range(biases[l].shape[0]):
            b_plus = np.copy(biases[l])
            b_plus[i] += epsilon
            biases_plus = biases[:]
            biases_plus[l] = b_plus
            output_plus = forward_propagation(X, weights, biases_plus, relu, sigmoid)
            loss_plus = mean_perceptron_loss(output_plus, y)

            b_minus = np.copy(biases[l])
            b_minus[i] -= epsilon
            biases_minus = biases[:]
            biases_minus[l] = b_minus
            output_minus = forward_propagation(X, weights, biases_minus, relu, sigmoid)
            loss_minus = mean_perceptron_loss(output_minus, y)

            b_grad[i] = (loss_plus - loss_minus) / (2 * epsilon)

        gradients.append(W_grad)
        gradients_biases.append(b_grad)

    return gradients, gradients_biases

def compute_accuracy(predictions, labels):
    # Binary classification
    predictions = predictions > 0.5  
    return np.mean(predictions == labels)


# Function for visualizing the final accuracies you get
def visualization(train_l, test_l, train_a, test_a):
    plt.figure(figsize=(15, 4))

    plt.subplot(1, 2, 1)
    plt.plot(train_a, label='Train accuracy', color="red")
    plt.plot(test_a, label='Test accuracy', color="green")
    plt.title('Mean perceptron loss over different epochs')
    plt.title('Classification accuracy over different epocs')
    plt.xlabel('Epochs')
    plt.ylabel('Mean Perceptron Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(train_l, label='Train loss', color="red")
    plt.plot(test_l, label='Test loss', color="green")
    plt.title('Mean perceptron loss over different epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    #plt.tight_layout()
    plt.show()
````

I also need to define a function that trains the MLP over different epoch values where we update our parameters with Adam optimizer,  and i print the resulting accuracies and losses for each epoch.

````python
def mlp(X_train, y_train, X_test, y_test, input_size, hidden_layer_num, hidden_layer_size, output_size, num_epochs=1000, learning_rate_init=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-8, hidden_activation=relu, output_activation=sigmoid, seed=1):
    np.random.seed(seed)
    weights, biases = initialize_weights_biases(input_size, hidden_layer_num, hidden_layer_size, output_size)
    params = weights + biases
    optimizer = AdamOptimizer(params, learning_rate_init=learning_rate_init, beta_1=beta_1, beta_2=beta_2, epsilon=epsilon)

    train_losses, train_accuracies, test_losses, test_accuracies = [], [], [], []
    start_time = time.time()

    for epoch in range(num_epochs):
        
        gradients, gradients_biases = compute_gradient(X_train, y_train, weights, biases)
        optimizer.update_params(params, gradients + gradients_biases)
        split_index = len(weights)
        weights = params[:split_index]
        biases = params[split_index:]

        train_output = forward_propagation(X_train, weights, biases, hidden_activation, output_activation)
        train_loss = mean_perceptron_loss(train_output, y_train)
        train_losses.append(train_loss)
        train_accuracy = compute_accuracy(train_output, y_train)
        train_accuracies.append(train_accuracy)

        test_output = forward_propagation(X_test, weights, biases, hidden_activation, output_activation)
        test_loss = mean_perceptron_loss(test_output, y_test)
        test_losses.append(test_loss)
        test_accuracy = compute_accuracy(test_output, y_test)
        test_accuracies.append(test_accuracy)
        
        # Printing the accuracies and losses
        if epoch % 100 == 0:  
            print(f"EPOCH: {epoch}")
            print(f"Train Loss: {train_loss:.5f}, Test Loss: {test_loss:.5f}.")
            print(f"Train Accuracy:: {train_accuracy:.5f}, Test Accuracy: {test_accuracy:.5f}.")
            print(f"-------------------------------------")

    training_time = time.time() - start_time
    print(f"Time for computation: {training_time:.2f}s.")
    
    return weights, biases, train_losses, train_accuracies, test_losses, test_accuracies, training_time
````
